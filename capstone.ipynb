```python
import pandas as pd

csv = 'Budget efficiency in Indonesia, February 2025.csv'
# df = pd.read_csv("/content/drive/MyDrive/Capstone/Budget efficiency in Indonesia, February 2025.csv", sep=';', header=None)

df = pd.read_csv(csv, sep=';', header=None)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df.columns = [
    "conversation_id_str", "created_at", "favorite_count", "full_text",
    "id_str", "image_url", "in_reply_to_screen_name", "lang", "location",
    "quote_count", "reply_count", "retweet_count"
]
df.columns = df.iloc[0]
df = df[1:]  # Hapus baris pertama
df = df.reset_index(drop=True) # Reset index dan bersihkan tipe data
```

```python
# Ubah tipe data kolom yang sesuai
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
numeric_cols = ['favorite_count', 'quote_count', 'reply_count', 'retweet_count']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

```

```python
from langchain_community.llms import Replicate
import os
# from google.colab import userdata
```

```python
from dotenv import load_dotenv
import os

# Ubah ini jika file .env ada di folder lain
env_path = os.path.join(os.getcwd(), '.env')  # path lengkap ke file .env

print("Looking for .env at:", env_path)  # Debug print

# Load environment dari .env
load_dotenv(dotenv_path=env_path)

# Ambil token
api_token = os.getenv('REPLICATE_API_TOKEN')

# Debug print nilai token
print("Token loaded:", api_token)

# Cek jika token masih None
if api_token is None:
    raise ValueError("REPLICATE_API_TOKEN not found. Pastikan .env file-nya benar dan berada di direktori kerja!")

# Set environment (opsional, jika memang diperlukan untuk package lain)
os.environ['REPLICATE_API_TOKEN'] = api_token
```

```python
# MODEL SETUP
model = 'ibm-granite/granite-3.3-8b-instruct'
```

```python
llm = Replicate(
    model = model,
    replicate_api_token = api_token
)
```

```python
ai_output = llm.invoke('hello')

print(ai_output)
```

```python
df_sample = df[['full_text']].sample(n=100, random_state=42)
```

```python
df_sample = df_sample.dropna()
df_sample = df_sample.apply(lambda x: x[:256])  # batasi max 256 karakter per teks
```

```python
from langchain_experimental.agents.agent_toolkits.pandas.base import (
    create_pandas_dataframe_agent
)

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df_sample,
    verbose=True,
    allow_dangerous_code=True,  # opt-in for Python REPL execution
)
```

```python
import nltk
nltk.download('punkt')
```

```python
agent.invoke({
    "input": "can cleaning the data and make new data frame, make all words to lowwercase "
})
```
